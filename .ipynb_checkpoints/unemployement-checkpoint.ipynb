# import the libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# read the data from csv file

```python
#read the csv file
unemploy= pd.read_csv('Unemployment in India.csv')
unemploy.head()
```

```python
unemploy.tail()
```

```python
unemploy.info()
```

```python
unemploy.describe()
```

```python
unemploy.isna().sum()
```

```python
#drop the null rows
unemploy.dropna(axis = 0, inplace = True)
```

```python
unemploy.isna().sum()
```

# Data Visualization

```python
plt.figure(figsize=(20, 10))
sns.heatmap(unemploy.corr().abs(),  annot=True)
```

```python
sns.countplot(x=unemploy['Area'])
plt.show()
```

```python
unemploy.columns
```

```python
unemploy.rename(columns={' Date':'Date',' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)',' Region':'Region'},inplace=True)
unemploy['Date']=pd.to_datetime(unemploy['Date'])
```

```python
#get day,month and time from Date
unemploy['day']=unemploy['Date'].dt.day
unemploy['month']=unemploy['Date'].dt.month_name()
unemploy['year']=unemploy['Date'].dt.year     
```

```python
unemploy.plot(kind="scatter", x='Estimated Unemployment Rate (%)', y='year')
plt.ylim(2018,2023)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Unemployment Rate (%)', hue="Region", data=unemploy)
plt.show()
```

```python
unemploy.plot(kind="scatter",  x='Estimated Unemployment Rate (%)', y='Region',marker ="^")
```
