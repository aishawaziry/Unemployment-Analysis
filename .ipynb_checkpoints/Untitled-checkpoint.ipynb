```python
unemploy.rename(columns={' Date':'Date',' Estimated Unemployment Rate (%)':'Estimated Unemployment Rate (%)'},inplace=True)
unemploy['Date']=pd.to_datetime(unemploy['Date'])
```

```python
unemploy['day']=unemploy['Date'].dt.day
unemploy['month']=unemploy['Date'].dt.month_name()
unemploy['year']=unemploy['Date'].dt.year     
```

```python

unemploy.plot(kind="scatter", x='Estimated Unemployment Rate (%)', y='year')
plt.show()
     
```
